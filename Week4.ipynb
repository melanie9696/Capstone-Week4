# IBM Data Science Capstone Project: Finding Optimal Locations for Opening Up a Bubble Tea Shop in Toronto Using K-means Clustering

## Introduction: Business Problem

The goal of this project is to find the optimal locations for a new bubble tea shop in the city of Toronto, Canada using the Foursquare location data. This report is targeted to potential investors who are interested in openning up a bubble tea shop in Toronto. Bubble tea, a sweer drink originally from Taiwan, has become exceedingly popular in North Ameriaca with a growing market. Toronto, the most populous city in Canada, has a large asian population of around 20 to 30 percent of all, indicating a potentially huge market for asian cusine. Only the choice of location in the business plan of opening a bubble tea shop will be explored. And areas where desert stores and asian restaurants are located will be targeted in consideration of customer base for bubble tea.
